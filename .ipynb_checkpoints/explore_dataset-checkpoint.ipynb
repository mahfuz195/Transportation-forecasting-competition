````python
import pandas as pd
import matplotlib.pyplot as plt
import helper
````

### Read data from file

````python
df=pd.read_csv("data/oct/1/xian/gps_20161001",
               delimiter=",", 
               names = ['driver_id','order_id','timestamp','longitude','latitude'])
````

### Filter data by location boundary
```
Point 1: 34.241, 108.943

Point 2: 34.241, 108.9415

Point 3: 34.234, 108.9415

Point 4: 34.234, 108.943
```

````python
df = df[df['longitude']>=108.9415]
df = df[df['longitude']<=108.943]
df = df[df['latitude']>=34.2329]  #34.2411
df = df[df['latitude']<=34.2422]  #34.2339
````

````python
#print (df.head(5)['longitude'])
#print (df.head(5)['latitude'])
#print (type(df['longitude']))
df.plot.scatter(x = 'longitude', y = 'latitude');
````

### Calculate speed, distance and direction of vehicles

````python
df = df.sort_values(by=['driver_id', 'timestamp'])
# Group the sorted dataframe by ID, and grab the initial value for lat, lon, and time.
df['lat0'] = df.groupby('driver_id')['latitude'].transform(lambda x: x.iat[0])
df['lon0'] = df.groupby('driver_id')['longitude'].transform(lambda x: x.iat[0])
df['t0'] = df.groupby('driver_id')['timestamp'].transform(lambda x: x.iat[0])
````

#### Apply Helper Functions

````python
# create a new column for distance
df['dist_km'] = df.apply(
    lambda row: helper.getDistanceFromLatLonInKm(
        lat1=row['latitude'],
        lon1=row['longitude'],
        lat2=row['lat0'],
        lon2=row['lon0']
    ),axis=1)

# create a new column for velocity
df['velocity_kmph'] = df.apply(
    lambda row: helper.calc_velocity(
        dist_km=row['dist_km'],
        time_start=row['t0'],
        time_end=row['timestamp']),axis=1)

df['bearing'] = df.apply(
    lambda row: helper.calc_bearing(
        lat1=row['lat0'],
        lon1=row['lon0'],
        lat2=row['latitude'],
        lon2=row['longitude']
    ),axis=1)
````

### Filter again interms of latitude and direction of vehicles

````python

df = df[df['latitude']>=34.234]  #34.2411
df = df[df['latitude']<=34.241]  #34.2339
df = df[df['velocity_kmph']!=0]

df_NB = df[df['bearing']<180.0]
df_SB = df[df['bearing']>180.0]

print(df_NB[['driver_id', 'timestamp', 'latitude', 'longitude', 'dist_km', 'velocity_kmph', 'bearing']].head(10))
print(df_SB[['driver_id', 'timestamp', 'latitude', 'longitude', 'dist_km', 'velocity_kmph', 'bearing']].head(10))
````

### Some Visualization of data

````python
import matplotlib.pyplot as plt
#drivers = df_NB.drop_duplicates('driver_id')

count = 0
for key, grp in df_NB.groupby(['driver_id']):
    fig, ax = plt.subplots() 
    ax = grp.plot(ax=ax, kind='line', x='timestamp', y='velocity_kmph', label=key)
    plt.legend(loc='best')
    plt.title('Speed profile of driver : ' + str(key))
    plt.show()
    count+=1
    if(count==4):
        break
````
